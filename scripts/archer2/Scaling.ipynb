```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
strong = pd.read_csv('data/strong_scaling_645327.csv')
weak = pd.read_csv('data/weak_scaling_645325.csv')
```

# Weak Scaling

```python
weak.keys()
```

### Points Vs Runtime

```python
# Calculate performance metrics
fig, ax = plt.subplots()

n_points = 1e6

ax.set_xscale('log')
ax.plot(weak['n_processes']*n_points, weak['runtime'],  '-o', label='Total Runtime')
ax.plot(weak['n_processes']*n_points, weak['sorting_time'], '-o', label='Sorting Time')
ax.plot(weak['n_processes']*n_points, weak['encoding_time'], '-o', label='Encoding Time')

ax.set_ylabel('Runtume (ms)')
ax.set_xlabel('Number of Points')

plt.legend()
plt.show()
```

### Points Vs Leaves

```python
fig, ax = plt.subplots()

ax.set_xscale('log')
ax.plot(weak['n_processes']*1e6, weak['n_leaves'], '-o')

ax.set_ylabel('Number of Leaves')
ax.set_xlabel('Number of Points')

plt.show()
```

### Leaves vs Runtime

```python
fig, ax = plt.subplots()

ax.set_xscale('log')

ax.plot(weak['n_leaves']*1e6, weak['runtime'], '-o', label='Total Runtime')
ax.plot(weak['n_leaves']*1e6, weak['sorting_time'], '-o', label='Sorting Time')

ax.set_ylabel('Runtime (ms)')
ax.set_xlabel('Number of Leaves')

plt.legend()
plt.show()
```
